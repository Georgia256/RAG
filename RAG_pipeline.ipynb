```python
from typing import Literal, Any
import logging
import PyPDF2
import sys
sys.path.append('/content/drive/MyDrive/RAG-llamaindex')
from logconf import log_setup
import os
os.environ['LOGGING_CONFIG'] = '/content/drive/MyDrive/RAG-llamaindex/logging.yaml'
from langchain_experimental.text_splitter import SemanticChunker
from math import ceil

log_setup()


logger = logging.getLogger("raft")
import json

DocType = Literal["api", "pdf", "json", "txt"]

def get_chunks(
    file_path: str,
    doctype: DocType = "pdf",
    chunk_size: int = 512,
    model: str = None
) -> list[str]:
    """
    Takes in a `file_path` and `doctype`, retrieves the document, breaks it down into chunks of size
    `chunk_size`, and returns the chunks.
    """
    chunks = []

    logger.info(f"Retrieving chunks from {file_path} of type {doctype}")

    if doctype == "api":
        with open(file_path) as f:
            api_docs_json = json.load(f)
        chunks = list(api_docs_json)
        chunks = [str(api_doc_json) for api_doc_json in api_docs_json]

        for field in ["user_name", "api_name", "api_call", "api_version", "api_arguments", "functionality"]:
            if field not in chunks[0]:
                raise TypeError(f"API documentation is not in the format specified by the Gorilla API Store: Missing field `{field}`")

    else:
        if doctype == "json":
            with open(file_path, 'r') as f:
                data = json.load(f)
            text = data["text"]
        elif doctype == "pdf":
            text = ""
            with open(file_path, 'rb') as file:
                reader = PyPDF2.PdfReader(file)
                num_pages = len(reader.pages)
                for page_num in range(num_pages):
                    page = reader.pages[page_num]
                    text += page.extract_text()
        elif doctype == "txt":
            with open(file_path, 'r') as file:
                data = file.read()
            text = str(data)
        else:
            raise TypeError("Document is not one of the accepted types: api, pdf, json, txt")

        num_chunks = ceil(len(text) / chunk_size)
        logger.info(f"Splitting text into {num_chunks} chunks using the {model} model.")

        #embed_model = HuggingFaceEmbedding(model_name=model)
        #text_splitter = SemanticChunker(embed_model, number_of_chunks=num_chunks)
        #chunks = text_splitter.create_documents([text])
        #chunks = [chunk.page_content for chunk in chunks]
        text_splitter = SentenceSplitter(chunk_size=chunk_size)#, chunk_overlap=200)
        chunks= text_splitter.get_nodes_from_documents(documents, show_progress=True)


    return chunks
```

```python
def strip_str(s: str) -> str:
    """
    Helper function for helping format strings returned by GPT-4.
    """
    l, r = 0, len(s)-1
    beg_found = False
    for i in range(len(s)):
        if s[i].isalpha():
            if not beg_found:
                l = i
                beg_found = True
            else:
                r = i
    r += 2
    return s[l:min(r, len(s))]
```

```python
def llama_generate_instructions_gen(chunk: Any, x: int = 5) -> list[str]:
    """
    Generates `x` questions / use cases for `chunk`. Used when the input document is of general types
    `pdf`, `json`, or `txt`.
    """

    # Define the prompt for the model
    prompt = (
        f"You are a synthetic question-answer pair generator. Given a chunk of context about some topic(s), "
        f"generate {x} example questions a user could ask and would be answered using information from the chunk. "
        f"For example, if the given context was a Wikipedia paragraph about the United States, an example question "
        f"could be 'How many states are in the United States?'\n\n"
        "The questions should be able to be answered in a few words or less. Include only the questions in your response. Do not include any prologue.\n\n"
        f"{chunk}"
    )

    # Generate questions using the specified model
    response = query_engine.query(prompt)

    # Print the response to inspect its structure
   # print(str(response))

    # Assuming the response is a dictionary with a 'text' key containing the model's output
    #if isinstance(str(response), dict) in str(response):
    queries = str(response).split('\n')
    queries = [q.strip() for q in queries if q.strip()]
    queries = [q for q in queries if any(c.isalpha() for c in q)]
    #else:
    #queries = []

    return queries


```

```python
#import groq

#instruction for cot

def encode_question_gen(question: str, chunk: Any) -> str:
    """
    Encode multiple prompt instructions into a single string for the general case (`pdf`, `json`, or `txt`).
    """

    prompts =[]

    prompt = f"""
    Question: {question}\n
    Context: {chunk}\n
    Answer this question using the information given in the context above. Here are things to pay attention to:
    - First provide step-by-step reasoning on how to answer the question.
    - In the reasoning, if you need to copy paste some sentences from the context, include them in ##begin_quote## and ##end_quote##. This would mean that things outside of ##begin_quote## and ##end_quote## are not directly copy-pasted from the context.
    - End your response with a final answer in the form <ANSWER>: $answer, the answer should be succinct.
    You MUST begin your final answer with the tag "<ANSWER>:".
    """

    prompts.append("You are a helpful question answerer who can provide an answer given a question and relevant context.")
    prompts.append(prompt)

    return prompts

# Example usage:
# question = "What is the capital of France?"
# chunk_of_text = "France is a country in Europe. The capital of France is Paris."
# prompt = encode_question_gen(question, chunk_of_text)
# print(prompt)
```

```python
from typing import List
import random
import datasets
from datasets import Dataset, load_dataset

def add_chunk_to_dataset(
    chunks: List[str],
    chunk: str,
    doctype: str = "general",
    x: int = 3,
    num_distract: int = 3,
    p: float = 0.8,
    model: str = 'llama3-8b-8192'
) -> None:
    """
    Given a chunk, create {Q, A, D} triplets and add them to the dataset.
    """
    global ds
    i = chunks.index(chunk)
    chunk = chunk.get_text()
    qs = llama_generate_instructions_gen(chunk, x)

    for q in qs:
        datapt = {
            "id": None,
            "type": None,
            "question": None,
            "context": None,
            "oracle_context": None,
            "cot_answer": None,
            "instruction": None
        }

        datapt["id"] = f"seed_task_{0 if not ds else ds.num_rows}"
        datapt["type"] = "api call" if doctype == "api" else "general"
        datapt["question"] = q

        # Add num_distract distractor docs
        docs = [chunk]
        indices = list(range(0, len(nodes)))
        indices.remove(i)
        for j in random.sample(indices, num_distract):
            docs.append(nodes[j].get_text())
        # Decides whether to add oracle document
        oracle = random.uniform(0, 1) < p
        if not oracle:
            docs[0] = nodes[random.sample(indices, 1)[0]].get_text()
        random.shuffle(docs)

        d = {
            "title": ["placeholder_title"]*(num_distract+1),
            "sentences": docs
        }

        # Ensure all elements are strings
        d["title"] = ["placeholder_title"] * (num_distract + 1)
        d["sentences"] = docs
        datapt["context"] = str(d)
        datapt["oracle_context"] = chunk

        prompt = encode_question_gen(q, chunk)

        # Convert the list to a single string
        prompt = ''.join(prompt)
        print(prompt)

        # Add answer to q
        datapt["cot_answer"] = generate_response(prompt)

        # Construct model instruction
        context = ""
        for doc in docs:
            context += "<DOCUMENT>" + str(doc) + "</DOCUMENT>\n"
        context += q
        datapt["instruction"] = context

        # Add to dataset
        if not ds:
            # Init ds
            ds_dict = {
                "id": [datapt["id"]],
                "type": [datapt["type"]],
                "question": [datapt["question"]],
                "context": [datapt["context"]],
                "oracle_context": [datapt["oracle_context"]],
                "cot_answer": [datapt["cot_answer"]],
                "instruction": [datapt["instruction"]]
            }
            ds = Dataset.from_dict(ds_dict)
        else:
            new_data = {
                "id": datapt["id"],
                "type": datapt["type"],
                "question": datapt["question"],
                "context": datapt["context"],
                "oracle_context": datapt["oracle_context"],
                "cot_answer": datapt["cot_answer"],
                "instruction": datapt["instruction"]
            }
            ds = ds.add_item(new_data)

def save_checkpoint(state, filename):
    with open(filename, 'w') as f:
        f.write(str(state))

def load_checkpoint(filename):
    with open(filename, 'r') as f:
        return int(f.read())
```
